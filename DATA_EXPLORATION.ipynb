# Task 2: Iris Dataset Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
# The Iris dataset is a classic and is readily available in seaborn's datasets.
iris_df = sns.load_dataset('iris')

print("Iris Dataset loaded successfully.")
```

## 1. Explore Dataset Structure

```python
# Display the first 5 rows of the dataset to get a glimpse of the data
display(iris_df.head())

# Display basic information about the dataset, including data types and non-null values
print("\nDataset Information:")
iris_df.info()

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(iris_df.describe())
```

## 2. Check for Missing Values

```python
# Check for any missing values across all columns
print("Missing values per column:")
display(iris_df.isnull().sum())

# Since there are no missing values, no handling is required for this dataset.
# If there were missing values, common strategies include:
# - Dropping rows/columns with missing data (if a small percentage).
# - Imputing missing values with mean, median, mode, or more advanced methods.
```

## 3. Visualize Data Distribution

### Univariate Analysis (Distribution of individual features)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of Iris Features', fontsize=16)

sns.histplot(iris_df['sepal_length'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(iris_df['sepal_width'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(iris_df['petal_length'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(iris_df['petal_width'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Bivariate Analysis (Features by Species)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Iris Features by Species', fontsize=16)

sns.boxplot(x='species', y='sepal_length', data=iris_df, ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length by Species')

sns.boxplot(x='species', y='sepal_width', data=iris_df, ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width by Species')

sns.boxplot(x='species', y='petal_length', data=iris_df, ax=axes[1, 0])
axes[1, 0].set_title('Petal Length by Species')

sns.boxplot(x='species', y='petal_width', data=iris_df, ax=axes[1, 1])
axes[1, 1].set_title('Petal Width by Species')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Multivariate Analysis (Pair Plot)

```python
# A pair plot visualizes pairwise relationships between variables.
# The 'hue' parameter colors points based on the 'species' column, helping to distinguish different classes.
print("Generating Pair Plot...")
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Features by Species', y=1.02) # Adjust title position
plt.show()
print("Pair Plot generated. This plot is excellent for observing correlations and species separability.")
```

### Correlation Heatmap

```python
# Calculate the correlation matrix for numerical features
correlation_matrix = iris_df.drop('species', axis=1).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

## 4. Advanced Visualizations and Statistical Summary

To further deepen our understanding of the Iris dataset, we'll introduce some more advanced visualization techniques and statistical summaries. These will help us to discern more subtle patterns and confirm visual observations with quantitative data.

### Violin Plots: Distribution and Density by Species

Violin plots combine aspects of box plots and kernel density plots. They show the distribution (density) of data across different categories (species) and also highlight quartiles and outliers, providing a richer view than simple box plots or histograms for comparison across groups.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Violin Plots of Iris Features by Species', fontsize=18)

sns.violinplot(x='species', y='sepal_length', hue='species', data=iris_df, ax=axes[0, 0], palette='pastel', legend=False)
axes[0, 0].set_title('Sepal Length Distribution by Species')

sns.violinplot(x='species', y='sepal_width', hue='species', data=iris_df, ax=axes[0, 1], palette='pastel', legend=False)
axes[0, 1].set_title('Sepal Width Distribution by Species')

sns.violinplot(x='species', y='petal_length', hue='species', data=iris_df, ax=axes[1, 0], palette='pastel', legend=False)
axes[1, 0].set_title('Petal Length Distribution by Species')

sns.violinplot(x='species', y='petal_width', hue='species', data=iris_df, ax=axes[1, 1], palette='pastel', legend=False)
axes[1, 1].set_title('Petal Width Distribution by Species')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Analysis of Violin Plots:

*   **Petal Length & Petal Width:** The violin plots for these features distinctly show that *setosa* has much shorter and narrower petals, with very little overlap with the other two species. *Versicolor* and *virginica* show more overlap but still maintain distinct distributions, especially in petal length where *virginica* tends to have longer petals.
*   **Sepal Length:** While there's more overlap, the distributions for sepal length show *setosa* generally having shorter sepals, *versicolor* in the middle, and *virginica* with longer sepals. The wider part of the violins indicates higher density, allowing us to see where most of the data points for each species lie.
*   **Sepal Width:** *Setosa* typically has wider sepals, followed by *virginica* and *versicolor*. The distributions are relatively normal for each species, though *setosa* shows a slightly bimodal tendency or a wider spread compared to *versicolor*.

### Joint Plots: Feature Relationships with Marginal Distributions

Joint plots are useful for visualizing the relationship between two variables (a scatter plot) and their individual distributions (histograms or KDEs on the margins). We can use this to examine how two features correlate within each species.

```python
plt.figure(figsize=(15, 6))

sns.jointplot(x='petal_length', y='petal_width', hue='species', data=iris_df, kind='kde', height=7)
plt.suptitle('Joint Plot of Petal Length vs. Petal Width by Species (KDE)', y=1.02)
plt.show()

sns.jointplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df, kind='scatter', height=7, marginal_kws=dict(fill=True))
plt.suptitle('Joint Plot of Sepal Length vs. Sepal Width by Species (Scatter)', y=1.02)
plt.show()
```

#### Analysis of Joint Plots:

*   **Petal Length vs. Petal Width:** The KDE joint plot clearly shows three distinct clusters corresponding to the three species. *Setosa* forms a tight, separate cluster in the lower-left corner (small petal length and width). *Versicolor* and *virginica* are also clustered, but with some overlap, showing a strong positive correlation between petal length and width within each group.
*   **Sepal Length vs. Sepal Width:** This scatter joint plot reveals more overlap between *versicolor* and *virginica* than seen with petal measurements. *Setosa* is still quite separable, typically having smaller sepal length but wider sepal width compared to the other two. The marginal histograms help to visualize the individual distributions along each axis for each species.

### Statistical Summary by Species

To quantify the differences observed in the visualizations, a grouped descriptive statistics table provides mean, standard deviation, and quartile information for each feature, broken down by species. This offers a numerical confirmation of the visual analysis.

```python
print("Descriptive Statistics of Iris Features Grouped by Species:")
display(iris_df.groupby('species').describe())
```

#### Analysis of Grouped Statistics:

*   The `describe()` output provides detailed statistics (count, mean, std, min, 25%, 50%, 75%, max) for each numerical feature, separated by species.
*   This table numerically confirms the visual observations: *setosa* has the smallest mean petal length and width, while *virginica* has the largest. *Versicolor* consistently falls in between.
*   Differences in standard deviation give insight into the variability within each species for each feature. For example, *setosa* generally shows less variability in petal dimensions compared to *versicolor* and *virginica*.
